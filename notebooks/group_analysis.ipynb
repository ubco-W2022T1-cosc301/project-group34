```python
import matplotlib.pyplot as plt

df = data.copy().describe()
df = df.filter(items=['Aquaculture production (metric tons)', 'Capture fisheries production (metric tons)'])

fig, ax = plt.subplots(figsize=(20, 7))
dfg= sns.barplot(data=df, ax=ax)
for item in dfg.get_xticklabels():
    item.set_rotation(45)
```

A bar chart of the mean production by aquaculture and capture method. Here we can see that the aquaculture production has a slightly higher portion than the one in capture fisheries production, roughtly about 15% higher, which may indicate that Aquaculture is still the main preference for these region to produce fish. Bare in mind that the Aquaculture is part of the sustainable method that reduces the impact of climate change. However, the capture fisheries still remains large, I assumed that this change of production only happens recently by the awareness of the climate change, or perhaps it may caused by many fishery production has not made a proper transition to aquaculture, or even many fish stock avaliable is not viable for aquaculture.
